```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')

n, bins,patches=plt.hist(iris['sepal_length'],
                         bins=20,
                         color='yellowgreen',
                         edgecolor='white')
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
plt.show()
```

```python
#iris 데이터셋 로드
import seaborn as sns
iris= sns.load_dataset("iris")
iris.info()
```

```python
# 상관관계 계수 계산
import pandas as pd
pd.set_option('display.precision',3)
corr_iris= iris[iris.columns[:-1].to_list()].corr()
corr_iris
```

```python
sns.heatmap( corr_iris,
            annot=True,
             cmap='coolwarm',
             square=True,
             linewidths=0.5,
             cbar_kws={'shrink':0.8});
```

```python
# mask 만들기
import numpy as np
mask= np.zeros_like(corr_iris)
mask
```

```python
np.triu_indices_from(mask)
```

```python
# 오른쪽 위 대각 행렬을 1로 바꾸기
mask[np.triu_indices_from(mask)]=1
mask
```

```python
mask_new=mask[1:,:-1]
mask_new
```

```python
corr_iris_new = corr_iris.iloc[1:, :-1]
sns.heatmap(corr_iris_new, annot=True, cmap='coolwarm', mask=mask_new,
            annot_kws={"size":10}, vmin=-1, vmax=1,
            cbar_kws={'shrink':.4},
            linewidths=.5);
```
